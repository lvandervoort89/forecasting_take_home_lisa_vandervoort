### Modeling

Prepare the data for modeling. Reset the index, rename the columns for Prophet, and drop unnecessary columns.

#### Separate Training Data

```python
# Reset index
train_data.reset_index(inplace=True)
```

```python
# Rename columns to match what Prophet expects 
train_data = train_data.rename(columns={'gas_price': 'y', 'month_and_year': 'ds'})

# Drop unnecessary columns for modeling
train_data = train_data.drop(['year', 'month'],  axis=1)
```

#### Separate Test Data

```python
# Rename columns to match what Prophet expects 
test_data = test_data.rename(columns={'gas_price': 'y', 'month_and_year': 'ds'})

# Drop unnecessary columns for modeling
test_data = test_data.drop(['year', 'month'],  axis=1)
```

#### Begin modeling. 
Start with baseline. Record each iteration and change. Detailed information about the rationale for decisions can be found in forecasting_take_home_lisa_vandervoort.pdf

```python
# Baseline model 

# Initiate model
model = Prophet()
    
# Fit the model to the training data
model.fit(train_data)
    
# Generate the future dataframe for monthly predictions 2 years out
future = model.make_future_dataframe(periods=24, freq='MS') 
    
# Forecast the future based on the model
forecast = model.predict(future)
    
# Plot the forecast
fig1 = model.plot(forecast)
    
# Plot the forecast of each component
fig2 = model.plot_components(forecast)
    
# Utilize built in cross_validation in Prophet 
cv_results = cross_validation(model, initial='730 days', period='180 days', horizon='365 days')
    
# Utilize built in performance_metrics in Prophet 
performance_results = performance_metrics(cv_results, metrics=['mae', 'rmse'])
    
# Average MAE and RMSE over the horizon (quick check of error)
mae_average_horizon = np.mean(performance_results.mae)
rmse_average_horizon = np.mean(performance_results.rmse)

print('Results from model are:\n'
          f'MAE mean for horizon: {mae_average_horizon},\n'
          f'RMSE mean for horizon: {rmse_average_horizon},\n')
```

```python
# Iteration #1: Add custom yearly seasonality 

# Initiate model, turn off automatic yearly_seasonality
model = Prophet(yearly_seasonality=False)

# Add custom yearly seasonality
model.add_seasonality(name='yearly', period=365.25, fourier_order=3, prior_scale=20, mode='additive')

# Fit the model to the training data
model.fit(train_data)
    
# Generate the future dataframe for monthly predictions 2 years out
future = model.make_future_dataframe(periods=24, freq='MS') 
    
# Forecast the future based on the model
forecast = model.predict(future)
    
# Plot the forecast
fig1 = model.plot(forecast)
    
# Plot the forecast of each component
fig2 = model.plot_components(forecast)
    
# Utilize built in cross_validation in Prophet 
cv_results = cross_validation(model, initial='730 days', period='180 days', horizon='365 days')
    
# Utilize built in performance_metrics in Prophet 
performance_results = performance_metrics(cv_results, metrics=['mae', 'rmse'])
    
# Average MAE and RMSE over the horizon (quick check of error)
mae_average_horizon = np.mean(performance_results.mae)
rmse_average_horizon = np.mean(performance_results.rmse)

print('Results from model are:\n'
          f'MAE mean for horizon: {mae_average_horizon},\n'
          f'RMSE mean for horizon: {rmse_average_horizon},\n')
```

```python
# Iteration #2: Adjust changepoint_prior_scale

# Initiate model, turn off auto yearly seasonality, adjust changepoint_prior_scale to be larger
model = Prophet(yearly_seasonality=False, changepoint_prior_scale=.1)

# Add custom yearly seasonality
model.add_seasonality(name='yearly', period=365.25, fourier_order=3, prior_scale=20, mode='additive')

# Fit the model to the training data
model.fit(train_data)
    
# Generate the future dataframe for monthly predictions 2 years out
future = model.make_future_dataframe(periods=24, freq='MS') 
    
# Forecast the future based on the model
forecast = model.predict(future)
    
# Plot the forecast
fig1 = model.plot(forecast)
    
# Plot the forecast of each component
fig2 = model.plot_components(forecast)
    
# Utilize built in cross_validation in Prophet 
cv_results = cross_validation(model, initial='730 days', period='180 days', horizon='365 days')
    
# Utilize built in performance_metrics in Prophet 
performance_results = performance_metrics(cv_results, metrics=['mae', 'rmse'])
    
# Average MAE and RMSE over the horizon (quick check of error)
mae_average_horizon = np.mean(performance_results.mae)
rmse_average_horizon = np.mean(performance_results.rmse)

print('Results from model are:\n'
          f'MAE mean for horizon: {mae_average_horizon},\n'
          f'RMSE mean for horizon: {rmse_average_horizon},\n')
```

```python
# Iteration #3: Adjust changepoint_range

# Initiate model, turn off auto yearly seasonality, adjust changepoint_prior_scale to be larger, and 
# adjust changepoint_range to be higher
model = Prophet(yearly_seasonality=False, changepoint_prior_scale=.1, changepoint_range=0.85)

# Add custom yearly seasonality
model.add_seasonality(name='yearly', period=365.25, fourier_order=3, prior_scale=20, mode='additive')

# Fit the model to the training data
model.fit(train_data)
    
# Generate the future dataframe for monthly predictions 2 years out
future = model.make_future_dataframe(periods=24, freq='MS') 
    
# Forecast the future based on the model
forecast = model.predict(future)
    
# Plot the forecast
fig1 = model.plot(forecast)
    
# Plot the forecast of each component
fig2 = model.plot_components(forecast)
    
# Utilize built in cross_validation in Prophet 
cv_results = cross_validation(model, initial='730 days', period='180 days', horizon='365 days')
    
# Utilize built in performance_metrics in Prophet 
performance_results = performance_metrics(cv_results, metrics=['mae', 'rmse'])
    
# Average MAE and RMSE over the horizon (quick check of error)
mae_average_horizon = np.mean(performance_results.mae)
rmse_average_horizon = np.mean(performance_results.rmse)

print('Results from model are:\n'
          f'MAE mean for horizon: {mae_average_horizon},\n'
          f'RMSE mean for horizon: {rmse_average_horizon},\n')
```

```python
# Create function to add additional regressor to model in next iteration
def is_great_recession(ds):
    '''
    A helper function that takes in a date and returns a value of 0 or 1 if the
    date is during the Great Recession (December 2007-June 2009).
    
    Parameters
    ----------
    date : The monthly gas average.
    
    Returns
    -------
    False if the month did not take place during the Great Recession and True if it did.
    ''' 
    day_to_check = pd.to_datetime(ds)
    if datetime(year=2007, month=12, day=1) <= day_to_check <= datetime(year=2009, month=6, day=1):
        return 1
    else:
        return 0

# Add column for great_recession   
train_data['great_recession'] = train_data['ds'].apply(is_great_recession)
test_data['great_recession'] = test_data['ds'].apply(is_great_recession)
```

```python
# Iteration #4: Add additional regressor for great recession months

# Initiate model, turn off auto yearly seasonality, adjust changepoint_prior_scale to be larger, and 
# adjust changepoint_range to be higher
model = Prophet(yearly_seasonality=False, changepoint_prior_scale=.1, changepoint_range=0.85)

# Add custom yearly seasonality
model.add_seasonality(name='yearly', period=365.25, fourier_order=3, prior_scale=20, mode='additive')

# Add additional regressor of great_recession
model.add_seasonality(name='great_recession', period=365.25, fourier_order=4, condition_name='great_recession')

# Fit the model to the training data
model.fit(train_data)
    
# Generate the future dataframe for monthly predictions 2 years out
future = model.make_future_dataframe(periods=24, freq='MS') 

# Add great recession column to future dataframe
future['great_recession'] = future['ds'].apply(is_great_recession)
    
# Forecast the future based on the model
forecast = model.predict(future)
    
# Plot the forecast
fig1 = model.plot(forecast)
    
# Plot the forecast of each component
fig2 = model.plot_components(forecast)
    
# Utilize built in cross_validation in Prophet 
cv_results = cross_validation(model, initial='730 days', period='180 days', horizon='365 days')
    
# Utilize built in performance_metrics in Prophet 
performance_results = performance_metrics(cv_results, metrics=['mae', 'rmse'])
    
# Average MAE and RMSE over the horizon (quick check of error)
mae_average_horizon = np.mean(performance_results.mae)
rmse_average_horizon = np.mean(performance_results.rmse)

print('Results from model are:\n'
          f'MAE mean for horizon: {mae_average_horizon},\n'
          f'RMSE mean for horizon: {rmse_average_horizon},\n')
```

### Final evaluation of the model on the test dataset

#### Fit final model

```python
# Concatenate train and test data together for final forecast
all_data = pd.concat([train_data, test_data], ignore_index=True)
```

```python
# Final model

# Initiate model, turn off auto yearly seasonality, adjust changepoint_prior_scale to be larger, and 
# adjust changepoint_range to be higher
model = Prophet(yearly_seasonality=False, changepoint_prior_scale=.1, changepoint_range=0.85)

# Add custom yearly seasonality
model.add_seasonality(name='yearly', period=365.25, fourier_order=3, prior_scale=20, mode='additive')

# Add additional regressor of great_recession
model.add_seasonality(name='great_recession', period=365.25, fourier_order=4, condition_name='great_recession')

# Fit the model to the training data
model.fit(train_data)
    
# Generate the future dataframe for monthly predictions 2 years out
future = model.make_future_dataframe(periods=24, freq='MS') 

# Add great recession column to future dataframe
future['great_recession'] = future['ds'].apply(is_great_recession)
    
# Forecast the future based on the model
forecast = model.predict(all_data)
    
# Plot the forecast
fig1 = model.plot(forecast)
    
# Plot the forecast of each component
fig2 = model.plot_components(forecast)
```

#### Evaluate final model

```python
# Get actual versus predicted values
y_actual = test_data.y
y_predicted = forecast.iloc[288:, -1]

# Assess on test data
test_mae = np.mean(np.abs(y_predicted - y_actual)) 
test_rmse = np.sqrt(mean_squared_error(y_actual, y_predicted))
```
